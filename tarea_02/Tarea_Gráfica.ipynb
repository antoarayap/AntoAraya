```python
#Números especiales del robot
# Imprime números del 1 al 100 que sean divisibles por 5 con mensajes especiales

for i in range(1, 101):
    # Primero verificamos si es divisible por 5
    if i % 5 == 0:
        # Si es divisible por 10, imprime "Bip Bip"
        if i % 10 == 0:
            print(f"{i} Bip Bip")
        # Si solo es divisible por 5, imprime "Bip"
        else:
            print(f"{i} Bip")
```

```python
#Detector de puntaje
# Solicita dos puntajes, los suma y clasifica el nivel del jugador

# Solicitamos los datos al usuario
puntaje1 = int(input("Ingresa primer puntaje: "))
puntaje2 = int(input("Ingresa segundo puntaje: "))

suma = puntaje1 + puntaje2

# Clasificación según el resultado de la suma
if suma < 50:
    print("Resultado: Nivel principiante")
elif 50 <= suma < 80:
    print("Resultado: Nivel intermedio")
else:
    print("Resultado: Nivel pro")
```

```python
#Calculador de perfil musical
# Solicitamos la edad (convertida a entero)
edad = int(input("¿Cuál es tu edad?: "))

# Para los gustos, comparamos la respuesta con "si" para obtener un True o False
# Usamos .lower() para que no importe si escriben "SI", "Si" o "si"
le_gusta_musica = input("¿Te gusta la música? (si/no): ").lower() == "si"
le_gusta_bailar = input("¿Te gusta bailar? (si/no): ").lower() == "si"

# Evaluación
if edad >= 18 and le_gusta_musica and le_gusta_bailar:
    print("Eres una estrella de la fiesta")
elif le_gusta_musica and not le_gusta_bailar:
    print("Te gusta la música, pero no tanto bailar")
else:
    print("Tienes tu propio estilo")
```
